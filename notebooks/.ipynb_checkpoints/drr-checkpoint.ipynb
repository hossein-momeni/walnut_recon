```python
import numpy as np
import torch

from diffdrr.data import read
from diffdrr.drr import DRR
from diffdrr.pose import convert
from diffdrr.visualization import plot_drr

import matplotlib.pyplot as plt
```

```python
subject_1 = read("walnut_data/walnut/data/Walnut1/gt.nii.gz")
subject_2 = read('walnut_data/walnut/data/Walnut2/gt_subsample_10x.nii.gz')
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
drr_1 = DRR(subject_1,
        sdd=199.006188,
        height=512,
        delx=0.074800
        # delx=1
          
).to(device)
```

```python
drr_2 = DRR(subject_2,
        sdd=199.006188,
        height=512,
        delx=0.074800
        # delx=1
          
).to(device)
```

```python
rotations = torch.tensor([[0.0, 0.0, 0.0]], device=device)
translations = torch.tensor([[0.0, 200.0, 0.0]], device=device)
img_1 = drr_1(rotations, translations, parameterization="euler_angles", convention="ZXY")
img_2 = drr_2(rotations, translations, parameterization="euler_angles", convention="ZXY")
plot_drr(torch.cat([img_1, img_2]))
```
